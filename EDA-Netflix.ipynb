### Find the oldest and newest released shows in the dataset:

```python
Netflix_Dataset['Release_Date'] = Netflix_Dataset['Release_Date'].fillna('1900-01-01')
Netflix_Dataset['Release_Date'] = pd.to_datetime(Netflix_Dataset['Release_Date'], errors='coerce')

oldest_show = Netflix_Dataset['Release_Date'].min()
newest_show = Netflix_Dataset['Release_Date'].max()

print("Oldest Show:", oldest_show)
print("Newest Show:", newest_show)
```

```python

avg_cast_per_category = Netflix_Dataset.groupby('Category')['Cast'].apply(lambda x: x.dropna().str.split(', ').apply(len).mean())
print(avg_cast_per_category)
```

```python
top_countries_unique_categories = Netflix_Dataset.groupby('Country')['Category'].nunique().sort_values(ascending=False).head(5)
print(top_countries_unique_categories)
```
